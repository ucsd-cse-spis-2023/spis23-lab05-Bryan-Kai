```python
from PIL import Image

# open the bear image and stores a reference to it in the bear variable
bear = Image.open("bear.png")

#view the image stored in the bear variable
display(bear)

# creating a function to output a grayscale image
def grayscale(im):
  (width, height) = im.size

  for x in range(width):
    for y in range(height):
      (red, green, blue) = im.getpixel((x, y))

      redLuminance = int(red * 0.21)
      greenLuminance = int(green * 0.72)
      blueLuminance = int(blue * 0.07)

      gray = int((redLuminance + greenLuminance + blueLuminance) / 3)



      im.putpixel((x, y), (gray, gray, gray))

grayscale(bear)
display(bear)
```
